```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}は{self.age}歳で、大人です")
        else:
            print(f"{self.name}は{self.age}歳で、大人ではありません")

humans = [
    Human("太郎", 15),
    Human("花子", 25),
    Human("健太", 18),
    Human("一郎", 30)
]

for human in humans:
    human.check_adult()
```
